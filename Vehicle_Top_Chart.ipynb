```python
import matplotlib.pyplot as plt

# Vehicle Count Distribution by Sides
# Response
data = {
    "message": "Vehicle found",
    "status": "success",
    "data": [
        {
            "context": "Allies",
            "total": 117
        },
        {
            "context": "Axis",
            "total": 75
        }
    ]
}

if data and data.get("status") == "success":
    # Dataset
    sides = [item["context"] for item in data["data"]]
    totals = [item["total"] for item in data["data"]]

    # Pie Chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        totals,
        labels=sides,
        autopct='%1.1f%%',
        startangle=140,
    )
    plt.title("Vehicle Count Distribution by Sides")
    plt.show()
else:
    print("No valid data to display!")
```
