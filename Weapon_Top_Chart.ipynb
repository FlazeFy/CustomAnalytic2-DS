```python
import matplotlib.pyplot as plt

# Weapon Count Distribution by Side
# Response
data = {
    "message": "Weapon found",
    "status": "success",
    "data": [
        {
            "context": "Axis",
            "total": 182
        },
        {
            "context": "Allies",
            "total": 203
        }
    ]
}

if data and data.get("status") == "success":
    # Dataset
    sides = [item["context"] for item in data["data"]]
    totals = [item["total"] for item in data["data"]]

    # Pie Chart
    plt.figure(figsize=(8,8))
    plt.pie(
        totals,
        labels=sides,
        autopct='%1.1f%%',
        startangle=140
    )
    plt.title("Weapon Count Distribution by Side")
    plt.show()
else: 
    print("No Valid data to show!")
```
