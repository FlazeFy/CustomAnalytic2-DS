```python
import matplotlib.pyplot as plt

# Ship Count Distribution by Side
# Response
data = {
    "message": "Ship found",
    "status": "success",
    "data": [
        {
            "context": "Axis",
            "total": 1154
        },
        {
            "context": "Allies",
            "total": 632
        }
    ]
}

if data and data.get("status") == "success":
    # Dataset
    sides = [item["context"] for item in data["data"]]
    totals = [item["total"] for item in data["data"]]

    # Pie Chart
    plt.figure(figsize=(8,8))
    plt.pie(
        totals,
        labels=sides,
        autopct='%1.1f%%',
        startangle=140,
    )
    plt.title("Ship Count Distribution by Sides")
    plt.show()
else:
    print("No valid data to display")
```
